```python
#HOW MANY MALES VS FEMALES APPLIED
import matplotlib.pyplot as plt

df['GENDER'].value_counts().plot(kind='bar', title='Gender Distribution', color=['lightpink', 'lightblue'])
plt.ylabel("Number of Applicants")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig("charts/gender_distribution.png", dpi=300, bbox_inches='tight')     #save chart to file
plt.show()
```

```python
#WHICH COURSES WERE APPLIED TO THE MOST?
df['APPLIED COURSE'].value_counts().plot(kind='barh', figsize=(8,10), title='Most Popular First Choice Courses')
plt.xlabel("Number of Applicants")
plt.ylabel("Course")
plt.tight_layout()
plt.savefig("charts/popular_courses.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#WHICH COURSES WERE APPLIED TO THE LEAST?
df['APPLIED COURSE'].value_counts(ascending=True).head(10).plot(kind='barh', figsize=(8,5), title='Least Popular First Choice Courses')
plt.xlabel("Number of Applicants")
plt.tight_layout()
plt.savefig("charts/least_popular_courses.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
#WHICH GENDER PREFERS WHICH COURSE?
import seaborn as sns

pivot = pd.crosstab(df['APPLIED COURSE'], df['GENDER'])
plt.figure(figsize=(10, 12))
sns.heatmap(pivot, annot=True, fmt='d', cmap='coolwarm')
plt.title("Course Preference by Gender")
plt.ylabel("Course")
plt.xlabel("Gender")
plt.tight_layout()
plt.show()
```

```python
#POPULARITY CATEGORIES
course_counts = df['APPLIED COURSE'].value_counts()
def categorize_course(course):
    count = course_counts.get(course, 0)
    if count >= course_counts.quantile(0.75):
        return "Popular"
    elif count <= course_counts.quantile(0.25):
        return "Niche"
    else:
        return "Moderate"

df['COURSE CATEGORY'] = df['APPLIED COURSE'].apply(categorize_course)
```
